```python
# Cargamos las librerías

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn
```

```python
# Cargamos el dataset train.csv

df = pd.read_csv("src\\datasets\\train.csv")
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
# Busca valores nulos
df.isnull().sum()
```

```python
# Busca valores extraños
columns = df.columns
n_values = [df[a].unique() for a in df.columns]

count = pd.DataFrame()
count['features'] = columns
count['n_values'] = n_values
count
```

```python
# Divido en numéricas y categóricas, para imputar los valores nulos por separado
categorical_cols = df.select_dtypes(include=['object']).columns
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns
```

```python
# Relleno con la media en variables numéricas
for col in numerical_cols:
    df[col] = df[col].fillna(df[col].mean())

# Relleno con la moda (el valor más frecuente) en variables categóricas
for col in categorical_cols:
    df[col] = df[col].fillna(df[col].mode()[0]) 
```

```python
# Comprobamos que se haya hecho bien la imputacion
df.isnull().sum()
df.info()
```

```python
# Dividimos la columna Cabin en 3 columnas, ya que representa deck/num/side y será más fácil 
# al entrenar el modelo y al hacer la conversion a numerico
df[["Deck", "Cabin_num", "Side"]] = df["Cabin"].str.split("/", expand=True)
try:
    df = df.drop('Cabin', axis=1)
except KeyError:
    print("Field does not exist")

# Convertir 'Cabin_num' a numérico
df["Cabin_num"] = pd.to_numeric(df["Cabin_num"], errors='coerce')
```

```python
# Borramos las columnas PassengerId y Name ya que no dan información relevante para el modelo
df = df.drop(['PassengerId', 'Name'], axis=1) 
```

```python
# Separar características (X) y etiquetas (y). (y) es la variable a predecir y (X) las predictoras
X = df.drop(columns=['Transported'])
y = df['Transported']
```

```python
X = pd.get_dummies(X)
```

```python
from sklearn.preprocessing import StandardScaler

# Definimos las variables numericas que queremos escalar
numeric_cols = ['Age', 'RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck', 'Cabin_num']

# Creamos el escalador
scaler = StandardScaler()

# Escalamos solo las columnas numéricas
X[numeric_cols] = scaler.fit_transform(X[numeric_cols])
```

```python
from sklearn.model_selection import train_test_split

# Dividir los datos en conjuntos de entrenamiento (70%) y prueba (30%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Verificar las formas de los datos
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
# Regresión logística
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score

# Dividir los datos en conjuntos de entrenamiento (70%) y prueba (30%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Verificar las formas de los datos
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)

# Definir los hiperparámetros para Logistic Regression
param_grid_log = {
    'C': [0.01, 0.1, 1.0, 10.0],    # Regularización
    'solver': ['liblinear', 'lbfgs'] # Solvers disponibles
}

# Crear el modelo base
model_log = LogisticRegression(max_iter=1000)

# Implementar GridSearchCV para ajustar los hiperparámetros
grid_search_log = GridSearchCV(estimator=model_log, param_grid=param_grid_log, cv=5, n_jobs=-1, verbose=1)

# Entrenar el modelo con los mejores hiperparámetros encontrados
grid_search_log.fit(X_train, y_train)

# Ver los mejores parámetros encontrados
print(f"Best hyperparameters for Logistic Regression: {grid_search_log.best_params_}")

# Predecir en el conjunto de prueba con el mejor modelo
y_pred_model_log = grid_search_log.best_estimator_.predict(X_test)

# Evaluar el modelo
accuracy_model_log = accuracy_score(y_test, y_pred_model_log)
print(f"Logistic Regression Accuracy (after hyperparameter tuning): {accuracy_model_log:.4f}")

# Validación cruzada con 5 particiones (folds) usando el mejor modelo
cv_scores_log = cross_val_score(grid_search_log.best_estimator_, X, y, cv=5)

# Imprimir los resultados promedio
print(f"Cross-validation scores (Logistic Regression): {cv_scores_log}")
print(f"Mean cross-validation accuracy (Logistic Regression): {cv_scores_log.mean():.4f}")
```

```python
# Random Forest
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score

# Dividir los datos en conjuntos de entrenamiento (70%) y prueba (30%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Definir los hiperparámetros para Random Forest
param_grid_rf = {
    'n_estimators': [100, 200, 300],     # Número de árboles
    'max_depth': [None, 10, 20, 30],     # Profundidad máxima del árbol
    'min_samples_split': [2, 5, 10],     # Mínimo número de muestras para dividir un nodo
    'min_samples_leaf': [1, 2, 4]        # Mínimo número de muestras en una hoja
}

# Crear el modelo base
model_rf = RandomForestClassifier(random_state=42)

# Implementar GridSearchCV para ajustar los hiperparámetros
grid_search_rf = GridSearchCV(estimator=model_rf, param_grid=param_grid_rf, cv=5, n_jobs=-1, verbose=1)

# Entrenar el modelo con los mejores hiperparámetros encontrados
grid_search_rf.fit(X_train, y_train)

# Ver los mejores parámetros encontrados
print(f"Best hyperparameters for Random Forest: {grid_search_rf.best_params_}")

# Predecir en el conjunto de prueba con el mejor modelo
y_pred_model_rf = grid_search_rf.best_estimator_.predict(X_test)

# Evaluar el modelo
accuracy_model_rf = accuracy_score(y_test, y_pred_model_rf)
print(f"Random Forest Accuracy (after hyperparameter tuning): {accuracy_model_rf:.4f}")

# Validación cruzada con 5 particiones (folds) usando el mejor modelo
cv_scores_rf = cross_val_score(grid_search_rf.best_estimator_, X, y, cv=5)

# Imprimir los resultados promedio
print(f"Cross-validation scores (Random Forest): {cv_scores_rf}")
print(f"Mean cross-validation accuracy (Random Forest): {cv_scores_rf.mean():.4f}")
```

```python
# XGBoost
import xgboost as xgb
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score

# Dividir los datos en conjuntos de entrenamiento (70%) y prueba (30%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Definir los hiperparámetros para XGBoost
param_grid_xgb = {
    'n_estimators': [100, 200],               # Número de árboles
    'max_depth': [3, 6, 10],                  # Profundidad máxima de los árboles
    'learning_rate': [0.01, 0.1, 0.2],        # Tasa de aprendizaje
    'subsample': [0.6, 0.8, 1.0],             # Fracción de muestras usadas para entrenar cada árbol
    'colsample_bytree': [0.6, 0.8, 1.0]       # Fracción de características usadas en cada árbol
}

# Crear el modelo base
xgb_model = xgb.XGBClassifier(random_state=42, eval_metric="logloss")

# Implementar GridSearchCV para ajustar los hiperparámetros
grid_search_xgb = GridSearchCV(estimator=xgb_model, param_grid=param_grid_xgb, cv=5, n_jobs=-1, verbose=1)

# Entrenar el modelo con los mejores hiperparámetros encontrados
grid_search_xgb.fit(X_train, y_train)

# Ver los mejores parámetros encontrados
print(f"Best hyperparameters for XGBoost: {grid_search_xgb.best_params_}")

# Predecir en el conjunto de prueba con el mejor modelo
y_pred_model_xgb = grid_search_xgb.best_estimator_.predict(X_test)

# Evaluar el modelo
accuracy_model_xgb = accuracy_score(y_test, y_pred_model_xgb)
print(f"XGBoost Accuracy (after hyperparameter tuning): {accuracy_model_xgb:.4f}")

# Validación cruzada con 5 particiones (folds) usando el mejor modelo
cv_scores_xgb = cross_val_score(grid_search_xgb.best_estimator_, X, y, cv=5)

# Imprimir los resultados promedio
print(f"Cross-validation scores (XGBoost): {cv_scores_xgb}")
print(f"Mean cross-validation accuracy (XGBoost): {cv_scores_xgb.mean():.4f}")
```
